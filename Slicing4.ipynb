```python
import pandas as pd

grades = pd.read_csv("School_Grades.csv")
grades.head()
```

```python
# Selecting specific rows
grades.loc[[1,2]]
```

```python
# Selecting specific rows and columns
grades.loc[[1,2,3],['FirstName','Subject','Grade']] 
```

```python
# Selecting specific rows and a range of columns
grades.loc[[1,2,3],'FirstName':'Grade'] 
```

```python
# Selecting a range of rows and columns
grades.loc[1:3,'FirstName':'Grade'] 
```

```python
# One condition and all columns

grades.loc[(grades['Marks']> 90),] 
```

```python
# One condition and specific columns
grades.loc[(grades['Marks']> 90),['FirstName', 'LastName', 'Grade']] 
```

```python
# Multiple conditions and specific columns

grades.loc[(grades['Marks']> 70) & (grades['Marks']< 80),['FirstName', 'LastName', 'Grade']] 
```

```python
# Multiple conditions and all columns

grades.loc[(grades['Marks']> 70) & (grades['Marks']< 80),] 
```
